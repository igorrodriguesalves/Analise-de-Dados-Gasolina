## 8. Qual o preço médio da gasolina em agosto de 2008?

```python
gas_comum[gas_comum['ANO-MÊS'] == "2008-08"]["PREÇO MÉDIO REVENDA"].mean()
#Pegamos a variável de gasolina que criamos, pegamos somente o mês e ano e depois chamamos a função de média
```

## 9. Qual o preço médio da gasolina em maio de 2014 em São Paulo?

```python
gas_comum[(gas_comum['ANO-MÊS'] == "2014-05") & (gas_comum["ESTADO"] == "SAO PAULO")]["PREÇO MÉDIO REVENDA"].mean()
#Primeiro fazemos o filtro, depois passamos a variável a ser mostrada e sua função
```

## 10. Quais estados a gasolina passou a valor de R$5,00 e quando isso ocorreu?

```python
gas_comum[["ESTADO", "ANO-MÊS", "PREÇO MÉDIO REVENDA"]][gas_comum["PREÇO MÉDIO REVENDA"] > 5].head()
```

## 11. Qual a média de preço dos estados da região sul em 2012?

```python
df_aux = gas_comum[gas_comum["DATA FINAL"].apply(lambda x: x.year) == 2012]
df_aux[df_aux["REGIÃO"] == "SUL"]["PREÇO MÉDIO REVENDA"].mean()
```

## 12.Obtenha uma tabela com a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
Tab_final["MES"] = Tab_final["DATA FINAL"].apply(lambda x: x.month)
#criamos a coluna somente dos meses a partir da coluna data final
df_Rio = Tab_final[Tab_final["ESTADO"] == "RIO DE JANEIRO"]
#Criamos uma variável somente com os dados do Rio de Janeiro
```

```python
df_mes_rio = df_Rio.groupby("ANO-MÊS")[['PREÇO MÉDIO REVENDA', 'MES']].last()
#Criamos outra variável que tem somente a Coluna de ANO-Mês 
```

```python
df_mes_rio[df_mes_rio['MES'] == 1]
#Com isso podemos ver a variação do preço da gasolina de cada mês mudando os anos
```

```python
df_mes_rio[df_mes_rio['MES'] == 1].count()
```

```python
(df_mes_rio[df_mes_rio['MES'] == 12] / df_mes_rio[df_mes_rio['MES'] == 12].shift(1) - 1) * 100
```

## DESAFIO: Crie uma tabela contendo uma série temporal com a diferença absoluta e percentual e entre os valores mais baratos e mais caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados

```python
df_max = Tab_final.groupby("ANO-MÊS").max()["PREÇO MÉDIO REVENDA"]
df_min = Tab_final.groupby("ANO-MÊS").min()["PREÇO MÉDIO REVENDA"]

idx_max = Tab_final.groupby("ANO-MÊS")["PREÇO MÉDIO REVENDA"].idxmax()
idx_min = Tab_final.groupby("ANO-MÊS")["PREÇO MÉDIO REVENDA"].idxmin()
```

```python
df_diferenca = pd.DataFrame()
```

```python
df_diferenca["abs_diff"] = df_max - df_min
```

```python
df_diferenca
```

```python
df_diferenca["percent_diff"] = ((df_max - df_min)/ df_min) * 100
```

```python
df_diferenca["percent_diff"]
```

```python
idx_max
```

```python
df_diferenca["ESTADO_MAX"] = Tab_final.loc[idx_max]["ESTADO"].values
df_diferenca["ESTADO_MIN"] = Tab_final.loc[idx_min]["ESTADO"].values
```
